```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
%matplotlib inline
plt.rcParams['figure.figsize']=(10,8)
```

```python
df=pd.read_csv('/content/movies.csv')
df.head()
```

```python
#check for missing data
for col in df.columns:
  pct_missing = np.mean(df[col].isnull())
  print(pct_missing)
```

```python
#columns datatypes
df.dtypes
```

```python
df['budget']=df['budget'].fillna(0)
df['gross']=df['gross'].fillna(0)
```

```python
#data cleaning
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
```

```python
df
```

```python
#create correct year column
df['correctyear']=df['year'].astype('str').str[:4]
df
```

```python
#sort values by gross column
df=df.sort_values(by='gross',inplace=False,ascending=False)
```

```python
#drop duplicate values
df.drop_duplicates()
df
```

```python
#find correlation between data

#Scatter plot of budget vs gross revenue
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget per film')
plt.show()
```

```python
df.head()
```

```python
#plot budget vs gross using seaborn
sns.regplot(x='budget',y='gross',data=df,scatter_kws={'color':'red'},line_kws={'color':'blue'})
```

```python
#to show correlation
df.corr(numeric_only=['True'])
```

```python
#plot correlation matrix
corr_matrix=df.corr(numeric_only=['True'])
sns.heatmap(corr_matrix,annot=True)
plt.show()
```

```python
df_numerized=df
for col_name in df_numerized.columns:
  if (df_numerized[col_name].dtype == 'object'):
    df_numerized[col_name] = df_numerized[col_name].astype('category')
    df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
#correlation matrix for df_numerized
corr_matrix=df.corr(numeric_only=['True'])
sns.heatmap(corr_matrix,annot=True)
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget per film')
plt.show()
```

```python
#check correlation for df_numerized
df_numerized.corr(numeric_only = [True])
```

```python
#use unstacking to check correlation pairs
corr_mat = df_numerized.corr(numeric_only = [True])
corr_pairs = corr_mat.unstack()
corr_pairs
```

```python
#check for sorted_pairs
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
#check for correlation higher than 0.5
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```

With this we can see the votes and budget have the highest correlation to gross earnings.

Company has low correlation.
